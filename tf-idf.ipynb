```python
import pandas as pd

df= pd.read_csv('datos/movie_metadata.csv')

df.head()
```

```python
df['genres']=df['genres'].str.replace('|', ' ') 
df
```

```python
df['plot_keywords']=df['plot_keywords'].str.replace('|', ' ') 
df
```

```python
df['texto']=df['genres']+' '+df['plot_keywords']
```

```python
row=df[['genres','plot_keywords','texto']].iloc[0]
print(row)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity, linear_kernel, euclidean_distances

```

```python
tfidf= TfidfVectorizer(max_features=2000)
```

```python
X=tfidf.fit_transform(df['texto'].fillna(''))
```

```python
peliculas = pd.Series(df.index, index=df['movie_title'])
```

```python
peliculas
```

```python
peliculas.index=peliculas.index.str.strip()

peliculas
```

```python
indice=peliculas['Avatar']
indice
```

```python
consulta = X[indice]
consulta
```

```python
print(consulta)
```

```python
consulta.toarray()
```

```python
similitud=cosine_similarity(X,consulta)
```

```python
similitud
```

```python
similitud[0,0]
```

```python
similitud=similitud.flatten()
```

```python
import matplotlib.pyplot as plt
plt.plot(similitud)
```

```python
(-similitud).argsort()
```

```python
plt.plot(similitud[(-similitud).argsort()])
plt.show()
```
